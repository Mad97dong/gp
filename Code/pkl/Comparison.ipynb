```python
import numpy as np
from IPython.display import clear_output
from numpy import linalg as LA
%matplotlib inline
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d.axes3d import Axes3D
import pickle
```

```python
list_ts = []
list_gd = []

for i in range(1, 6):
    file_name = "./thompson_Ackley" + str(i) + ".pkl" 
    open_file = open(file_name, "rb")
    loaded_list = pickle.load(open_file)
    open_file.close()
    list_ts.append(loaded_list)
    
    file_name = "./GD_Ackley" + str(i) + ".pkl" 
    open_file = open(file_name, "rb")
    loaded_list = pickle.load(open_file)
    open_file.close()
    list_gd.append(loaded_list)
```

```python
pl.figure(figsize=(10, 7), dpi=80)
pl.plot(np.mean(list_ts, axis=0), 'b', lw=3, label='ts')
pl.plot(np.mean(list_gd, axis=0), 'r', lw=3,  label='GD')

pl.xlabel("cost")
pl.ylabel("function value")
pl.grid(True)
pl.legend()
# pl.ylim([0, 5])
```
